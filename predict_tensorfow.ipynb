```python
from tensorflow.keras.models import load_model
from PIL import Image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
import os
import time
```

```python
labels = ['black', 'blue', 'brown', 'green', 'orange', 'pink', 'red', 'silver', 'white', 'yellow']
exp_th = 1
```

```python
model = load_model('models/best_model_color_20_04_2_0.9663.hdf5')
```

```python
import os 
import shutil
path_img_crop = 'D:/PROJECTS/AI/Color/Data_vehicle/images_test/crops2'
folder_crop = os.listdir(path_img_crop)
path_save = 'D:/PROJECTS/AI/Color/Data_vehicle/images_test/test'
i = 0
for folder in folder_crop:
    path_sub_folder = os.path.join(path_img_crop, folder)
    images = os.listdir(path_sub_folder)
    for image in images:
        shutil.move(os.path.join(path_sub_folder,image), os.path.join(path_save,f'image_crop2{i}.jpg'))
        i+=1
```

```python
path_img_test = 'D:/PROJECTS/AI/Color/Data_vehicle/images_test/runs/exp_test_new2/red'
images_test = os.listdir(path_img_test)
path_save = 'D:/PROJECTS/AI/Color/Data_vehicle/images_test/runs'
path_save += f'/exp{exp_th}'
exp_th += 1
for image in images_test:
    img_ori = Image.open(os.path.join(path_img_test, image))
    img = img_ori.crop(( 0,  0, img_ori.width -  0, img_ori.height -  0))
    img = img.resize((224, 224)) # Resize the image to match your model's input size
    img_array = np.array(img)
    img_array = img_array.astype('float32') / 255.0 # Normalize pixel values to [0, 1]
    img_array = np.expand_dims(img_array, axis=0) 
    start_time = time.time()
    predictions = model.predict(img_array)
    end_time = time.time()
    total_time = (end_time - start_time) * 1000  # Convert to milliseconds
    print(f"Total predict time: {total_time:.2f} ms")
    # Get the predicted class index and class name
    predicted_class_index = np.argmax(predictions[0])

    folder_save = os.path.join(path_save, labels[predicted_class_index])
    if not os.path.exists(folder_save):
        os.makedirs(folder_save)
    img_ori.save(os.path.join(folder_save,image))
```
